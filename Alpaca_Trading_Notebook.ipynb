### Using Colab Secrets for Secure API Key Storage

Storing sensitive information like API keys directly in your notebook code is not recommended due to security risks. Colab provides a secure way to store and access your secrets (like API keys) using its built-in 'Secrets' feature.

**Benefits of using Colab Secrets:**
1.  **Security:** Your API keys are not directly exposed in the notebook, preventing accidental commits to version control or unauthorized access.
2.  **Convenience:** Once stored, secrets can be easily accessed in any Colab notebook without re-entering them.
3.  **Best Practice:** Aligns with industry best practices for handling credentials.

**How to set up Colab Secrets:**
1.  On the left sidebar of your Colab notebook, click the 'ðŸ”‘ Secrets' icon.
2.  Click '+ New secret'.
3.  For `Name`, enter `APCA_API_KEY_ID` and for `Value`, paste your Alpaca API Key ID.
4.  Click '+ New secret' again.
5.  For `Name`, enter `APCA_API_SECRET_KEY` and for `Value`, paste your Alpaca Secret Key.
6.  Make sure the 'Notebook access' toggle is enabled for your current notebook.
